```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Импортируем библиотеки
import pandas as pd
```

```python
# Загружаем набор данных
ECommerce_Shipping_Data = pd.read_csv('/content/drive/MyDrive/ECommerce_Shipping_Data.csv')
ECommerce_Shipping_Data.head()
```

```python
# Сохраняем в новый фрейм данных столбцы предикторов
X=ECommerce_Shipping_Data.drop(['Reached_on_Time_Y_N'], axis=1)
# Сохраняем в переменную y отклик
y=ECommerce_Shipping_Data.Reached_on_Time_Y_N
# Преобразуем категориальные переменные в dummy
X=pd.get_dummies(X, drop_first=True)
```

```python
# Проведем разбиение исходных наблюдений предикторов и отклика
# на обучающую выборку и тестовую
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
X_test1 = X_test
X_test1['knn'] = knn_pred
X_test1['Reached_on_Time_Y_N'] = y_test
X_test1.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, knn_pred)
```

```python
X_test1.groupby(['knn', 'Reached_on_Time_Y_N'])['knn'].count().unstack()
```
