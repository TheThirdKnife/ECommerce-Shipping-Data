Разделение на матрицу признаков и зависимую переменную

```python
full_X = imputed_df.drop(['Продукция_растениеводства'], axis=1)
full_y = imputed_df['Продукция_растениеводства']
```

```python
full_X['Искусственные_данные'] = df.isna().any(axis=1) # Метка об обработке null-значений
```

```python
# Создадим временные ряды, используя информацию за последние три года
time_window = 3
rows = len(full_y.index) - time_window
```

```python
timeline_cols = {}

for i in range(time_window):
  column_name = f'Продукция_растениеводства-{time_window - i}'
  timeline_cols[column_name] = full_y.iloc[i:rows+i].to_list()
timeline_cols['Продукция_растениеводства'] = full_y.iloc[time_window:].to_list()

timeline_df = pd.DataFrame(timeline_cols)
time_X = timeline_df.drop(['Продукция_растениеводства'], axis=1)
time_y = timeline_df['Продукция_растениеводства']
```

```python
timeline_cols = {}

for i in range(time_window):
  datarows_before = full_X.iloc[i:rows+i]
  for column in datarows_before.columns:
    column_name = f'{column}-{time_window - i}'
    timeline_cols[column_name] = datarows_before[column].to_list()
  column_name = f'Продукция_растениеводства-{time_window - i}'
  timeline_cols[column_name] = full_y.iloc[i:rows+i].to_list()
timeline_cols['Продукция_растениеводства'] = full_y.iloc[time_window:].to_list()

full_timeline_df = pd.DataFrame(timeline_cols)
full_time_X = full_timeline_df.drop(['Продукция_растениеводства'], axis=1)
full_time_y = full_timeline_df['Продукция_растениеводства']
```

```python
corr = full_timeline_df.corr()
# Отберём лучшие 15 признаков по их корреляции
best_cols = corr.abs().nlargest(16, 'Продукция_растениеводства').index

truncated_df = full_timeline_df[best_cols]
trunc_X = timeline_df.drop(['Продукция_растениеводства'], axis=1)
trunc_y = timeline_df['Продукция_растениеводства']
best_cols[1:]
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

experiment_number = 0
min_mse, min_mse_model = 1000000000, None
max_r2, max_r2_model = 0, None


def experiment(*method_name):
  global experiment_number
  experiment_number += 1
  print(f'Эксперимент №{experiment_number}:', *method_name)
def rescaled(model):
  return make_pipeline(StandardScaler(), model)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split


def train_test_score(model, data_X, data_y):
  global max_r2, min_mse, max_r2_model, min_mse_model
  train_X, test_X, train_y, test_y = train_test_split(data_X, data_y, test_size=0.15, random_state=69)
  model.fit(train_X, train_y)
  pred_y = model.predict(test_X)
  print('Результат эксперимента:')
  # Больше - лучше
  if r2_score(test_y, pred_y) > max_r2: max_r2, max_r2_model = r2_score(test_y, pred_y), model
  print(f'Коэффициент детерминации: {r2_score(test_y, pred_y):.4f}')
  # Ошибка, меньше - лучше
  if mean_squared_error(test_y, pred_y) < min_mse: min_mse, min_mse_model = mean_squared_error(test_y, pred_y), model
  print(f'Среднеквадратичная ошибка (MSE): {mean_squared_error(test_y, pred_y):.4f}')
```

```python
def train_test_model(model, model_name: str):
  experiment(model_name, 'на исходных данных')
  train_test_score(model, full_X, full_y)
  experiment(model_name, 'на временных рядах')
  train_test_score(model, time_X, time_y)
  experiment(model_name, 'на временных рядах данных')
  train_test_score(model, full_time_X, full_time_y)
  experiment(model_name, 'на уменьшенных данных')
  train_test_score(model, trunc_X, trunc_y)
```

```python
from sklearn.linear_model import LinearRegression

train_test_model(LinearRegression(), 'Линейная регрессия')
```

```python
from sklearn.linear_model import LassoCV

train_test_model(rescaled(LassoCV(max_iter=100000)), 'Линейная регрессия с линейными штрафами и стандартизацией данных')
```

```python
from sklearn.linear_model import RidgeCV

train_test_model(rescaled(RidgeCV()), 'Линейная регрессия с квадратичными штрафами и стандартизацией данных')
```

```python
from sklearn.neural_network import MLPRegressor

net_params = {'max_iter': 100000, 'solver': 'lbfgs'}
train_test_model(rescaled(MLPRegressor((50,100,50,),**net_params)), 'Нейронная сеть без градиентного спуска')
```
